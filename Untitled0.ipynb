```python
import tensorflow as tf
from tensorflow import keras

import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler

train_labels=[]
train_samples=[]

#5% younger people with symptoms and 5% old people without symptoms
for i in range(50):
  younger= randint(13,64)
  train_samples.append(younger)
  train_labels.append(1)

  older= randint(65,100)
  train_samples.append(older)
  train_labels.append(0)

# 95% old people with symptoms and 95% young people without symptoms
for i in range(1000):
  younger=randint(65,100)
  train_samples.append(younger)
  train_labels.append(0)

  older=randint(65,100)
  train_samples.append(older)
  train_labels.append(1)

train_labels=np.array(train_labels)
train_samples=np.array(train_samples)
train_labels,train_samples= shuffle(train_labels,train_samples)

scaler= MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))

for i in scaled_train_samples:
  print(i)
```
